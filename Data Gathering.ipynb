```python
from bs4 import BeautifulSoup
import pandas as pd

# Original HTML snippet
html_snippet = '''
<tr class="app" data-appid="1623730" data-cache="1707253748">
<td>1.</td>
<td class="applogo text-left"><a href="/app/1623730/charts/"><img alt="" src="https://cdn.cloudflare.steamstatic.com/steam/apps/1623730/capsule_sm_120.jpg?t=1707253748"/></a></td>
<td><a href="/app/1623730/charts/">Palworld</a></td>
<td data-sort="486221">486,221</td>
<td data-sort="944328">944,328</td>
<td data-sort="2101867">2,101,867</td>
<td class="add">+</td>
</tr>
'''

# Parse HTML
soup = BeautifulSoup(html_snippet, 'html.parser')

# Extract information
game_name = soup.find('td').find_next_sibling().find_next_sibling().text.strip()
app_id = soup.find('tr')['data-appid']
current_players = soup.find_all('td', {'data-sort': True})[0].text
hr_peak = soup.find_all('td', {'data-sort': True})[1].text
all_time_peak = soup.find_all('td', {'data-sort': True})[2].text

# Create DataFrame
data = {
    'Game Name': [game_name],
    'App ID': [app_id],
    'Current Players': [current_players],
    '24 Hr Peak': [hr_peak],
    'All-Time Peak': [all_time_peak]
}

df = pd.DataFrame(data)
print(df)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
# summary_list = []

# Create an empty list to store data from all snippets
all_data = []

for article in articles:
    # Parse HTML
    soup = BeautifulSoup(html_snippet, 'html.parser')

    # Extract information
    game_name = soup.find('td').find_next_sibling().find_next_sibling().text.strip()
    app_id = soup.find('tr')['data-appid']
    current_players = soup.find_all('td', {'data-sort': True})[0].text
    hr_peak = soup.find_all('td', {'data-sort': True})[1].text
    all_time_peak = soup.find_all('td', {'data-sort': True})[2].text

    # Store data in a dictionary
    data = {
        'Game Name': game_name,
        'App ID': app_id,
        'Current Players': current_players,
        '24 Hr Peak': hr_peak,
        'All-Time Peak': all_time_peak
    }

    # Append data to the list
    all_data.append(data)

# Create DataFrame from the list of dictionaries
df = pd.DataFrame(all_data)
print(df)
```
